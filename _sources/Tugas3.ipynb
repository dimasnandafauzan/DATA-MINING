## Implementasi K-NN (Tugas 3)

```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
plt.ion()
```

```python
datairisurl="https://raw.githubusercontent.com/Tegar11-F/Data-Set/fa20d41a8676343efb6cb833304d9fb54b807d67/IRIS.csv"
datairis=pd.read_csv(datairisurl)
```

```python
datairis.head(30)
```

```python
#periksa jumlah baris dan kolom dalam dataset
datairis.shape
```

```python
#buat kerangka data dengan semua data pelatihan kecuali kolom target
X = datairis.drop(columns=['species'])
#periksa apakah variabel target telah dihapus
X.head()
```

```python
#nilai target terpisah
y = datairis['species'].values
#view target values
y[0:29]
```

```python
from sklearn.model_selection import train_test_split
#bagi dataset menjadi data latih dan uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Buat pengklasifikasi KNN
knn = KNeighborsClassifier(n_neighbors = 3)
# Sesuaikan pengklasifikasi dengan data
knn.fit(X_train,y_train)
```

```python
#tampilkan 29 prediksi model pertama pada data uji
knn.predict(X_test)[0:29]
```

```python
#periksa keakuratan model kami pada data uji
knn.score(X_test, y_test)
```
