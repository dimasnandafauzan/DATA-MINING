```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
plt.ion()
```

```python
dataseturl="https://raw.githubusercontent.com/Tegar11-F/Data-Set/main/smoke_detection_iot.csv"
data=pd.read_csv(dataseturl)
```

```python
data.head(4)
```

```python
def diss (n,m):
  a = 2
  b = 0
  for col in ['Raw Ethanol','Pressure[hPa]']:
    if data[col][n] == data[col][m]:
      b+=1
  return (a-b)/a

print(diss(1,1))
print(diss(1,2))
print(diss(4,4))
```
